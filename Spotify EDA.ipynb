```python
# Import Packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataset

df_tracks = pd.read_csv('/Users/shashankuparwat/Desktop/Spotify EDA/tracks.csv')
df_artists = pd.read_csv('/Users/shashankuparwat/Desktop/Spotify EDA/artists.csv')
df_genre = pd.read_csv('/Users/shashankuparwat/Desktop/Spotify EDA/SpotifyFeatures.csv')
```

```python
df_tracks.head()
```

```python
# Check NULL Values

pd.isnull(df_tracks).sum()
```

```python
# df Info
df_tracks.info()
```

```python
# Least Popular Tracks

df_bot10 = df_tracks.sort_values('popularity', ascending = True).head(10)
df_bot10 
```

```python
# Top 10 Popular Tracks - Above 90 ratings

df_10popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
df_10popular[:10]
```

```python
# Descriptive Statistics

df_tracks.describe().transpose()
```

```python
# Change Index to Release Date

df_tracks.set_index("release_date", inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head(10)
```

```python
# Locate Artists

df_tracks[["artists"]].iloc[10]
```

```python
# Covert Duration(ms) to Seconds

df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
df_tracks.duration.head()
```

```python
# Correlation Map
    
df_corr = df_tracks.drop(["key", "mode", "explicit"], axis = 1).corr(method = 'pearson')
plt.figure(figsize = (14,6))

heatmap = sns.heatmap(df_corr, annot = True, fmt = ".1g", vmin = -1, vmax = 1, center = 0, cmap = "inferno", linewidth = 1, linecolor = "Black")
heatmap.set_title('Correlation Heatmap Between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
# Create  Sample Dataset

df_sample = df_tracks.sample(int(0.004*len(df_tracks)))

print(len(df_sample))
```

```python
# Regression Plot

plt.figure(figsize = (10, 6))
sns.regplot(data = df_sample, y = "loudness", x = "energy", color = "b").set(title = "Loudness VS Energy")
```

```python
plt.figure(figsize = (10, 6))
sns.regplot(data = df_sample, y = "popularity", x = "acousticness", color = "c").set(title = "Popularity VS Acousticness")
```

```python
# Dist Plot - Songs Per Year

# Create YEAR col from RELEASE_DATE col

df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete = True, aspect = 2, height = 5, kind = 'hist').set(title = "Number of songs per year")
```

```python
# Duration of songs over the years

total_dr = df_tracks.duration
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title = "Year VS Duration")
plt.xticks(rotation = 90)
```

```python
# Line Plot

total_dr = df_tracks.duration
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr, ax = ax).set(title = "Year VS Duration")
plt.xticks(rotation = 90)
```

```python
# Duration of songs for different Genres

plt.title("Duration of songs for different Genres")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y = 'genre', x = 'duration_ms' , data = df_genre)
plt.xlabel("Duration in milliseconds")
plt.ylabel("Genres")
```

```python
# Top 5 Genres

sns.set_style(style = "darkgrid")
plt.figure(figsize = (10, 5))
famous = df_genre.sort_values("popularity",ascending = False).head(10)
sns.barplot(y = 'genre', x = 'popularity' , data = famous).set(title = "Top 5 Genres by Popularity")
```
